# Task-1: Zero-Shot Prompting – Prime Number Check

# Prompt 1 Write a Python function to check whether a number is prime.


```python
def is_prime(n):
    for i in range(2, n):
        if n % i == 0:
            print("Not Prime")
            return
    print("Prime")

is_prime(7)
```

# Prompt 2 Write a Python function to check whether a number is prime.Handle numbers less than or equal to 1.


```python
def is_prime(n):
    if n <= 1:
        print("Not Prime")
        return
    for i in range(2, n):
        if n % i == 0:
            print("Not Prime")
            return
    print("Prime")

is_prime(1)
is_prime(9)
is_prime(11)
```

# Prompt 3 Write an efficient Python function to check whether a number is prime. Use correct logical conditions.

```python
def is_prime(n):
    if n <= 1:
        print("Not Prime")
        return
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            print("Not Prime")
            return
    print("Prime")

is_prime(2)
is_prime(4)
is_prime(17)
```

# Task-2: One-Shot Prompting – Sum of Elements in a List

# Prompt 1 Write a Python function to calculate the sum of elements in a list.

```python
def list_sum(arr):
    total = 0
    for i in arr:
        total += i
    print("Sum:", total)

list_sum([1, 2, 3, 4])
```

#Task-3: Few-Shot Prompting – Extract Digits from Alphanumeric String

#Prompt 1 Write a Python function to extract digits from an alphanumeric string.

```python
def extract_digits(s):
    result = ""
    for ch in s:
        if ch.isdigit():
            result += ch
    print(result)

extract_digits("a1b2c3")
```

# Task-4: Zero-Shot vs Few-Shot Prompting – Count Vowels

# Prompt 1 Write a Python function to count the number of vowels in a string.


```python
def count_vowels(s):
    count = 0
    for ch in s:
        if ch in "aeiou":
            count += 1
    print("Vowels:", count)

count_vowels("Education")
```

# Prompt 2 Write a Python function to count vowels in a string.

```python
def count_vowels(s):
    count = 0
    for ch in s.lower():
        if ch in "aeiou":
            count += 1
    print("Vowels:", count)

count_vowels("Education")
count_vowels("sky")
```

# Task-5: Few-Shot Prompting – Minimum of Three Numbers (Without min())

# Prompt 1 Write a Python function to find the minimum of three numbers without using the built-in min() function.

```python
def minimum(a, b, c):
    if a < b and a < c:
        print(a)
    elif b < a and b < c:
        print(b)
    else:
        print(c)

minimum(3, 7, 5)
minimum(10, 2, 8)
```

# Prompt 2 Write a Python function to find the minimum of three numbers without using min().

```python
def minimum(a, b, c):
    if a <= b and a <= c:
        print(a)
    elif b <= a and b <= c:
        print(b)
    else:
        print(c)

minimum(4, 4, 6)
minimum(2, 2, 2)
```
